```python
import pyshark
from manuf import manuf
import requests
import time

p = manuf.MacParser(update=False)
```

```python
# set up capture
capture = pyshark.LiveCapture(interface='en0', display_filter='wlan.fc.type_subtype == 0x0004', monitor_mode=True)
capture.sniff(timeout=10)
```

```python
# create lists
pcks = []
macs = []
```

```python
for pkt in capture:
    pcks.append(pkt)
    
    try:
        wlan = pkt.layers[3]
        wlan2 = pkt.layers[2]
        radio = pkt.layers[0]
        
        ssid = wlan.ssid
        vendor = p.get_all(wlan2.ta)
        mac = wlan2.ta
        
        if mac not in macs:
            macs.append(mac)
            print('\x1b[1;31m'+'New device'+'\x1b[0m'+' total {}'.format(len(macs)))
            #resp = requests.get('https://api.wigle.net/api/v2/network/search?ssid={}'.format(ssid), headers={'Authorization': 'Basic QUlEODQ4MWMwNWRiMGE4ZGQ1ZjU3ZjRlN2Q5YzI2ZDM2NzI6YmI0ZTkwYTM4NmNlYTI0NDJjY2RmMjMyNTdjMGIyMGU'})
            #print(resp.json())

        print(vendor.manuf, ssid, radio.dbm_antsignal, wlan2.ta)
    except:
        pass
```

```python
# parameters
signal_strength_threshold = -60
ignored_devices = []
new_devices_per_minute = 10
```

```python
seen_timestamps = []

def process_packet(packet):
    try:
        # check if mac is not in ignored devices list
        mac = pkt.layers[2].ta
        if any(mac in s for s in ignored_devices):
            return False

        
        # first add current timestampp
        seen_timestamps.append(int(time.time()))
        
        # then check how many packets were observed in the past min
        past_min = int(time.time())-60
        packets_in_last_min = 0
        for timestamp in seen_timestamps:
            if timestamp >= past_min:
                packets_in_last_min += 1
        
        #print('seen {} packets in last min'.format(packets_in_last_min))
        
        # now check if below or above threshold
        if packets_in_last_min >= new_devices_per_minute:
            print('\x1b[1;31m'+'TRIGGERED'+'\x1b[0m'+' {} devices seen in the past minute'.format(packets_in_last_min))
        
        
        # check if the signal strength is beyond threshold        
        signal_strength = int(pkt.layers[0].dbm_antsignal)
        if signal_strength > signal_strength_threshold:
            print('\x1b[1;31m'+'TRIGGERED'+'\x1b[0m'+' device seen {}dB away'.format(signal_strength))
            return True
        
        
        return False
    except:
        pass 
```
