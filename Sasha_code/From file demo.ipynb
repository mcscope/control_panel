```python
import pyshark
from manuf import manuf
import requests
p = manuf.MacParser(update=False)
```

```python
# set up capture
capture = pyshark.FileCapture(input_file="/Users/sasha/Dropbox/code/18_06_bm_art/180704_cli_capture.pcap", display_filter='wlan.fc.type_subtype == 0x0004')
```

```python
# create lists
pcks = []
macs = []
```

```python
for pkt in capture:
    pcks.append(pkt)
    
    try:
        wlan = pkt.layers[3]
        wlan2 = pkt.layers[2]
        radio = pkt.layers[0]
        
        ssid = wlan.ssid
        vendor = p.get_all(wlan2.ta)
        mac = wlan2.ta
        
        if mac not in macs:
            macs.append(mac)
            print('\x1b[1;31m'+'New device'+'\x1b[0m'+' total {}'.format(len(macs)))
            #resp = requests.get('https://api.wigle.net/api/v2/network/search?ssid={}'.format(ssid), headers={'Authorization': 'Basic QUlEODQ4MWMwNWRiMGE4ZGQ1ZjU3ZjRlN2Q5YzI2ZDM2NzI6YmI0ZTkwYTM4NmNlYTI0NDJjY2RmMjMyNTdjMGIyMGU'})
            #print(resp.json())

        print(vendor.manuf, ssid, radio.dbm_antsignal, wlan2.ta)
    except:
        pass
```
